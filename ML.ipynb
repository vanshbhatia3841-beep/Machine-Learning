```python
from sklearn.neighbors import KNeighborsClassifier

# 1. Prepare the Training Data
# Features [Hours Studied, Hours Slept]
X = [
    [2, 7], # A
    [3, 6], # B
    [4, 7], # C
    [6, 5], # D
    [7, 6], # E
    [8, 5]  # F
]

# Labels [0 = Fail, 1 = Pass]
y = [0, 0, 0, 1, 1, 1]

# 2. Initialize the KNN Model
# We use n_neighbors=3 to look at the 3 closest points
knn = KNeighborsClassifier(n_neighbors=3)

# 3. Train the model
knn.fit(X, y)

# 4. Predict for New Data (5 hours studied, 6 hours slept)
new_data = [[5, 6]]
prediction = knn.predict(new_data)

# Output the result
status = "Pass" if prediction[0] == 1 else "Fail"
print(f"Prediction for (5, 6): {status}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

# 1. Prepare Data
X = np.array([[2, 7], [3, 6], [4, 7], [6, 5], [7, 6], [8, 5]])
y = np.array([0, 0, 0, 1, 1, 1])
new_data = np.array([[5, 6]])

# 2. Train Model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
prediction = knn.predict(new_data)

# 3. Create a Mesh Grid for Decision Boundaries
h = .02  # Step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict across the entire grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 4. Plotting
plt.figure(figsize=(8, 6))

# Plot the decision boundary
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA']) # Red for Fail, Green for Pass
plt.pcolormesh(xx, yy, Z, cmap=cmap_light, shading='auto')

# Plot the training points
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='red', edgecolors='k', label='Fail (Train)')
plt.scatter(X[y == 1, 0], X[y == 1, 1], c='green', edgecolors='k', label='Pass (Train)')

# Plot the new data point
plt.scatter(new_data[0, 0], new_data[0, 1], c='blue', marker='*', s=200, label=f'New Point (Pred: {"Pass" if prediction[0] == 1 else "Fail"})')

plt.xlabel('Hours Studied')
plt.ylabel('Hours Slept')
plt.title('KNN Decision Boundary (k=3)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('Tennis.csv')

# Inspect the data
print(df.info())
print(df.head())
```

```python
import math

def calculate_entropy(data, target_col):
    values = data[target_col].value_counts()
    entropy = 0
    total = len(data)
    for count in values:
        p = count / total
        entropy -= p * math.log2(p)
    return entropy

def calculate_info_gain(data, feature, target_col):
    total_entropy = calculate_entropy(data, target_col)
    values = data[feature].unique()
    weighted_entropy = 0
    total = len(data)

    feature_info = []
    for val in values:
        subset = data[data[feature] == val]
        subset_entropy = calculate_entropy(subset, target_col)
        weighted_entropy += (len(subset) / total) * subset_entropy
        feature_info.append({
            'value': val,
            'count': len(subset),
            'entropy': subset_entropy
        })

    gain = total_entropy - weighted_entropy
    return gain, feature_info

target = 'play tennis'
features = [col for col in df.columns if col != target]

total_entropy = calculate_entropy(df, target)
print(f"Total Entropy (S): {total_entropy:.4f}")

gains = {}
details = {}
for feature in features:
    gain, info = calculate_info_gain(df, feature, target)
    gains[feature] = gain
    details[feature] = info
    print(f"Information Gain for {feature}: {gain:.4f}")

# Find best feature
best_feature = max(gains, key=gains.get)
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Encode labels
df_encoded = df.copy()
label_encoders = {}
for column in df_encoded.columns:
    le = LabelEncoder()
    df_encoded[column] = le.fit_transform(df_encoded[column])
    label_encoders[column] = le

X = df_encoded.drop('play tennis', axis=1)
y = df_encoded['play tennis']

# Initialize and train Decision Tree
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(X, y)

# Plot the tree
plt.figure(figsize=(12, 8))
plot_tree(clf,
          feature_names=X.columns,
          class_names=label_encoders['play tennis'].classes_,
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Decision Tree for Tennis Dataset (Entropy)")
plt.savefig('decision_tree.png')

# Show column mapping for reference
mapping = {col: list(label_encoders[col].classes_) for col in label_encoders}
print("Encoding Mapping:")
print(mapping)
```

```python
# New input data for prediction
new_data = pd.DataFrame({
    'Outlook': ['rainy'],
    'Temprature': ['mild'],
    'humidity': ['normal'],
    'windy': [True]
})

# Encode the new data using the fitted label encoders
new_data_encoded = new_data.copy()
for column, le in label_encoders.items():
    if column in new_data_encoded.columns:
        new_data_encoded[column] = le.transform(new_data_encoded[column])

# Make prediction
prediction_encoded = clf.predict(new_data_encoded.drop('play tennis', axis=1, errors='ignore'))

# Decode the prediction
prediction_label = label_encoders['play tennis'].inverse_transform(prediction_encoded)

print(f"Based on the conditions (Outlook: rainy, Temprature: mild, Humidity: normal, Windy: True), the prediction is: {prediction_label[0]}")
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load dataset
df = pd.read_csv('Tennis.csv')

# Remove spaces from column names (VERY IMPORTANT)
df.columns = df.columns.str.strip()

# Encode categorical columns
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])

# Check columns
print("Columns:", df.columns)

# Set target (CHANGE if needed)
target = df.columns[-1]   # last column assumed target

X = df.drop(target, axis=1)
y = df[target]

# Train model
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X, y)

# Plot tree
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=X.columns, filled=True)
plt.show()
```
